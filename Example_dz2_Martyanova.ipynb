# Домашнее задание - 2

Эффект переноса валютного курса в цены -- это увеличение общего уровня цен в результате ослабления национальной валюты
Например, если удешевление рубля на 10% приводит к инфляции в 3%, то говорят что эффект переноса равен 0,3. 

С конца 2014 года Банк России перешел к политике инфляционного таргетирования. Инфляционное таргетирование (ИТ) -- режим монетарной политики, соответствующий следующим принципам:
* Количественная цель: ЦБ публично оглашает целевой уровень инфляции
* Приоритет ценовой стабильности: Достижение этого уровня является главной целью ЦБ
* Прозрачность и подотчетность: ЦБ подробно объясняет общественности, как именно каждый из предпринятых им шагов помогает ему достичь главной цели

Неотъемлимой частью политики инфляционного таргетирования является режим плавающего валютного курса. 
Если эффект валютного переноса велик, то это колебания курса приводят к значительному росту инфляции. В этом случае политика инфляционного таргетирования сводится на нет. 

В данном домашнем задании необходимо ответить на следующий исследовательский вопрос: как изменилась величина эффекта переноса валютного курса в российские цены в 2015-2019 годах по сравнению с предшествующим периодом?

**Инфляция** -- устойчивое повышение уровня цен в экономике. Существует несколько показателей, отражащих динамику цен, например
* индекс цен производителей
* дефлятор ВВП
* индекс потребительских цен (ИПЦ)

ИПЦ измеряет изменение во времени стоимости набора продовольственных, непродовольственных, товаров и услуг, потребляемых средним домохозяйством. Я выбрала ИПЦ в качестве одного из показателей инфляции по нескольким причинам: 
* ИПЦ чаще всего выбирают многие исследователи в качестве показателя инфляции, так он отражает динамику стоимости жизни населения
* ИПЦ удобен из-за его месячной периодичности расчета и оперативной публикации
* и самое главное, Банк России устанавливает цель по инфляции именно  по ИПЦ. (подробнее: [основные цели КДП на сайте Банка России](http://www.cbr.ru/publ/ondkp/))

Попробуем оценить модель для данных, начиная с ноября 2014 (в ноябре Банк России перешел к политике таргетирования инфляции) по апрель 2019. 

* Индекс выпуска, скачанный с Росстата, представляет собой отношение уровня выпуска предыдущего периода к выпуску текущего периода, то есть это цепной индекс. Для того, чтобы получить базисный индекс (база -- февраль 2009), я поочередно перемножила предыдущий базисный индекс с последующим цепным индексом, то есть

$\dfrac{Y_t}{Y_0} = \dfrac{Y_t}{Y_{t-1}} \cdot \dfrac{Y_{t-1}}{Y_0}$

То есть столбец 'Индекс выпуска, базисный' -- это временной ряд, "пронормированный" на выпуск первого периода. 

* Я преобразовала ИПЦ в индекс прироста, так как, например, в скрипте SVAR, выложенном на on.econ, используется именно прирост ИПЦ.
* Взяла номинальный эффективный курс как логрифм, первые разности характеризуют темп прироста (как в задаче с файлом US_Macro)

```python
data2014 = data.loc['2014-11-01':].copy()
```

```python
data2014.head()
```

```python
cpi = data2014['ИПЦ, к предыдущему месяцу'].apply(lambda x: (x-100)/100)
y = data2014['Индекс выпуска, базисный']
er = np.log(data2014['Номинальный эффективный обменный курс рубля'])
brent = data2014['Цена на нефть марки Brent']
m2 = data2014['Денежный агрегат М2']
```

```python
variables = [brent, cpi, er, m2, y]
var_names = ['brent', 'cpi', 'er', 'm2', 'y']
```

```python
# функция для вывода результатов тестов на стационарность
def stat_tests(variables, var_names):
    for i, j in zip(variables, var_names):
        print('Дики-Фуллер p-value для {}: {}'.format(j, adfuller(i)[1]))
        print('KPSS p-value для {}: {}'.format(j, kpss(i)[1])) 
```

Выводы из графиков:
* ряд выпуска нестационарен: явно заметны тренд и сезонность, нужно взять первые разности и сделать коррекцию на сезонность.
* явно есть тренд у денежной массы, но ряд уже и так сезонно сглаженный. 

```python
fig, ax = plt.subplots(nrows=5, ncols=1, figsize=(10,10))
fig.subplots_adjust(hspace=0.4) #расстояние между графиками

ax[0].set_title('$cpi_{t}$', fontsize=10)
ax[0].plot(cpi)
ax[1].set_title('$ y_{t}$', fontsize=10)
ax[1].plot(y)
ax[2].set_title('$ er_{t}$', fontsize=10)
ax[2].plot(er)
ax[3].set_title('$ brent_{t}$', fontsize=10)
ax[3].plot(brent)
ax[4].set_title('$ m2_{t}$', fontsize=10)
ax[4].plot(m2);
```

Тесты отвергают нулевую гипотезу о стационарности всех рядов при любом разумном уровне значимости, поэтому далее рассмотрим первые разности.

```python
stat_tests(variables, var_names)
```

```python
# убираем сезонность
y = y - seasonal_decompose(y).seasonal
cpi = cpi - seasonal_decompose(cpi).seasonal
er = er - seasonal_decompose(er).seasonal
brent = brent - seasonal_decompose(brent).seasonal
```

```python
# первые разности
delta_cpi = np.diff(cpi)
delta_y = np.diff(y)
delta_er = np.diff(er)
delta_brent = np.diff(brent)
delta_m2 = np.diff(m2)

var_delta = [delta_brent, delta_cpi, delta_er, delta_m2, delta_y]
delta_names = ['delta_'+ i for i in var_names]
```

```python
fig, ax = plt.subplots(nrows=5, ncols=1, figsize=(10,10))
fig.subplots_adjust(hspace=0.4)
ax[0].set_title('$\Delta cpi_{t}$', fontsize=10)
ax[0].plot(delta_cpi);
ax[1].set_title('$\Delta y_{t}$', fontsize=10)
ax[1].plot(delta_y)
ax[2].set_title('$\Delta er_{t}$', fontsize=10)
ax[2].plot(delta_er)
ax[3].set_title('$\Delta brent_{t}$', fontsize=10)
ax[3].plot(delta_brent)
ax[4].set_title('$\Delta m2_{t}$', fontsize=10)
ax[4].plot(delta_m2);
```

```python
stat_tests(var_delta, delta_names)
```

# SVAR

Класс SVAR в питоне требует определить ограничения A и B. Изначально в пакете прописана форма:
$A y_t = A_1 y_{t-1} + \ldots + A_p y_{t-p} + B \varepsilon_t $. [github](https://github.com/statsmodels/statsmodels/blob/master/statsmodels/tsa/vector_ar/svar_model.py)

Мне было бы удобнее задавать структурные ограничения для уравнения в приведенной форме (то есть структурные шоки), поэтому я определяю только матрицу В. 

Структурные ограничения для краткосрочных шоков в виде декомпозации Холески: 

$ \begin{pmatrix}
u^{oil} \\ u^{cpi} \\ u^{neer} \\ u^{M2} \\ u^{GDP} \\
\end{pmatrix} $ =
$ \begin{pmatrix}
a_1 & 0 & 0 & 0 & 0 \\
b_1 & b_2 & 0 & 0 & 0\\
c_1 & c_2 & c_3 & 0 & 0 \\
d_1 & d_2 & d_3 & d_4 & d_0 \\
f_1 & f_2 & f_3 & f_4 & f_5 \\ 
\end{pmatrix} $ 
$\cdot$
$\begin{pmatrix}
\varepsilon^{oil} \\ \varepsilon^{cpi} \\ \varepsilon^{neer} \\ \varepsilon^{M2} \\ \varepsilon^{GDP} \\
\end{pmatrix} $

* предполагается, что цены на нефть определяются на мировом рынке, то есть экзогенны. Россия -- малая открытая экономика по экспорту нефти, поэтому происходящие внутри нее шоки не влияют на мировую цену.
* шок валютного курса зависит от цен на нефть (нефть занимает важное место в структуре российского экспорта).
* шок денежной массы зависит от обменного курса.
* шок выпуска зависит от почти всех остальных шоков, что следует из модели совокупного спроса и модели IS-LM-BP.
* шок инфляции в данном случае самая эндогенная переменная.

```python
dat_delta = pd.DataFrame(var_delta).T
dat_delta.columns = delta_names

B = np.array([['E', 0, 0, 0, 0],
              ['E', 'E', 0, 0, 0],
              ['E', 'E', 'E', 0, 0], 
              ['E', 'E', 'E', 'E', 0], 
              ['E', 'E', 'E', 'E', 'E']])
```

```python
bic = []
aic = []
lags = list(range(1, 7+1))
for lag in lags:
    model = VAR(dat_delta).fit(lag)
    bic.append(model.bic)
    aic.append(model.aic)
    
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(7, 5));
fig.subplots_adjust(hspace=0.4)
ax[0].set_title('BIC', fontsize=15)
ax[0].plot(lags, bic);
ax[1].set_title('AIC', fontsize=15)
ax[1].plot(lags, aic);
```

```python
# или автоматически определяем порядок регрессии
model = VAR(dat_delta)
a = model.select_order(7)
a.selected_orders
```

```python
model_macro = SVAR(endog = dat_delta, svar_type = 'B', B = B).fit(1)
print(model_macro.summary())
```

```python
dat_delta8 = pd.DataFrame([delta_cpi8, delta_y8, delta_er8, delta_brent8, delta_m28]).T
dat_delta8.columns = delta_names
#получается ситуацияБ аналогичная первой
model = VAR(dat_delta8)
a = model.select_order(7)
a.selected_orders
```

```python
model_macro8 = SVAR(endog = dat_delta8, svar_type = 'B', B = B).fit(1)
print(model_macro8.summary())
```
