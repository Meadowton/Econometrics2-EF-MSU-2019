# Влияет ли знание иностранных языков на доход?

# Введение и обзор литературы

Владение языками является важным элементом человеческого капитала. Изучение связи между благосостоянием индивида и его языковыми компетенциями является одним из основных направлений в economics of language. 

Цель данного проекта -- выявить связь между заработной индивида и его языковыми компетенциями. В данном проекте использовались данные российского мониторинга экономического положения и здоровья населения НИУ-ВШЭ за 2016 г. и 2012 г. 

## The Economics of Linguistic Diversity: How Many Languages Make Sense? Victor Ginsburgh, Shlomo Weber, 2011
В данной книге проведен обзор основных направлений economics of language:
1. индексы, отражающие этно-лингвистические дистанции, в том число их применение в эмпирических экономических исследованиях:
    * влияние языкового разнообразия на международную торговлю и миграцию
    * литературные переводы
    * оценка вероятности победы на Евровидении (языковые и культурные дистанции влияют и на поведение при голосовании
    * оценка вероятности этнических конфликтов

4. Языковая политика, в том числе подробно рассмотрена политика Европейского Союза с точки зрения издержек и disenfranchisement indices.

## Оценка выигрыша от знания языка
Проблема решения индивидов об изучении языка и соотношении их выгод и издержек интересует экономистов не только с точки зрения эмпирических результатов, но и с точки зрения теории игр ([Selten and Pool (1991)](http://econ.ucsb.edu/~tedb/Courses/Ec100C/seltenpool.pdf), [Church and King (1993)](https://www.jstor.org/stable/135911)).
[Ginsburgh, Ortu˜no-Ort´ın, Weber (2007)](https://e-archivo.uc3m.es/bitstream/handle/10016/5512/learning_ortuno_JEBO_2007_ps.pdf?sequence=1) не только разработали теоретико-игровую модель, но и попытались дать эмпирические оценки.

Большинство исследований посвящено влиянию знания языка на заработок мигрантов.

Двуязычные страны, главным образом Канада, а также, например, Украина и Казахстан, часто выступают объектом для подобных исследований.   

Во многих исследованиях были найдены значимые положительные эффекты от знания языков на заработок индивидов. Считается, что знание языка улучшает возможности трудоустройства. Однако очевидно, что далеко не все языки аналогичны в этом отношении.

Про оценке выигрыша индивида от знания иностранного языка возникает проблема эндогенности, как и при оценке выигрыша от образования. Знание языка зависит может зависеть от ненаблюдаемых исследователем характеристик, например, способностей к языкам. 

### ["Male-female earnings differentials and the role of language in Canada, Ontario, and Quebec"](https://www.jstor.org/stable/pdf/134804.pdf?casa_token=N6ydUHfV2KcAAAAA:zE5pAFP1809qPCOe9agt6DfGiCM7wkNuvb3ATt4S5gKWq6IPxuogErYUf2U5ai5gGbiFiJZIUSQI9aQ8MWZ3A6BUdUs452EpFMRarmnKbckWy7Pg7lQ)
Shapiro и Stelcner не обнаружили значимого выигрыша от знания языка другой группы для носителей английского и французского языков. В последующей работе, основанной на данных переписи 1990 года они обнаружили значимый положительный выигрыш для носителей французского языка, знающих английский.

### ["Earnings by Language Group in Quebec in 1980 and Emigration from Quebec between 1976 and 1981", Grenier (1987)](https://www.jstor.org/stable/pdf/135415.pdf?casa_token=gPIrg9crXNoAAAAA:yvCL2kZ6vAD1EeoB0dPsyErxfkBYGyn1ElT5-Shv0vREZKCVhUkpiUNm-bA9xlr470XTJNyonyDWhoGUYukm0O4rx2sXPA77q2lwobDaKglO0ahw5FA)
В данной статье анализируется миграция из канадской провинции Квебек в период с 1976 по 1981 год. Большую долю мигрироваших составляли носители английского языка. Так как решение мигрировать может быть связано с ожидаемым уровнем заработка, при простом сравнении средней заработной в 1976 и 1981 возникает проблема смещения из-за самоотбора. В данной работе использовалась система из двух одновременных уравнений с использованием выборки из мужчин, взятой из перепеси 1981 года. В качестве засисимой переменной в первом уравнении была выбрана бинарная переменная, равная 1, если индивид жил в Квебеке, 
регрессоры включали в себя:
* возраст
* бинарные переменные для знания языков и их использования
* семейного положения
* места происхождения и иммиграции
Во втором уравнении оценивался логарифм общего дохода индивида, в состав независимых переенных были включены:
* число лет образования
* опыт работы и квадрат опыта работы
* семейное положение
* логарифм количества отработанных недель в году
* бинарная перемнная для самозанятости
* бинарная перемееная, равная 1 для живущих в Монреале 
* аналогичные бинарные переменные для знания языков и их использования
Оценка регресии показала, что, при прочих равных, носитель английского языка, знающий французский, с большей вероятностью оставался в Квебеке. Также результаты показали, что для носителей французского нет проблемы самоотбора, в то время как для носителей английского языка она существует: наблюдалась корреляция между ошибками уравнений для веростности остаться и заработка.Таким образом, оставшиеся в Квебеке носители английского приняли такое решение, потому что они получали относительно высокий заработок по сравнению с теми, кто решил уехать. При прочих равных, билингвизм увеличивал заработную плату носителей французского на 4 п.п., носителей английского на 12 п.п. с контролем самоотбора, и на 8 п.п. -- без. 
Автор пришел к выводу, что смещение самоотбора действительно присутсвует и занижает отдачу от французского языка.

### [Immigrant access to work in Montreal and Toronto, Gilles Grenier and Serge Nadeau, 2011](http://www.cjrs-rcsr.org/V34/1/CJRS-RCSR-34-1-03.pdf)
В данной статье рассматривается проблема разрыва в уровнях занятости между мигрантами и местными в Торонто и Монреале. В Монреале развыв значительно выше, так как мигранты вынуждены нести большие издержки приспособления: единственным официальным языком является французский и больше половины населения владеет в той или иной мере обоими языками. 

Авторы использует данные переписи 2006 года (the Statistics Canada 2006 Census Microdata Master File). Они оценивают вероятность быть занятым на полный рабочий день весь год с помощью линейной модели. В качестве регрессоров использовались:
1. для всех наблюдений:
    * опыт работы
    * уровень образования 
    * язык, на котором говорят дома
    * семейное положение
    * знание официальных языков
    * другие доходы в семье
    * наличие детей
    * бинарная переменная, если индивид посещал школу за последние 9 месяцев
2. для иммигрантов
    * наличие гражданства
    * место получения наивсшей ступени образования
    * место рождения
    * количество лет в иммиграции

В данной работе использовался метод декомпозиции [Blinder–Oaxaca](https://en.wikipedia.org/wiki/Blinder%E2%80%93Oaxaca_decomposition), поэтому авторы отдели предпочтение линейной регрессии. Всего были оценены 4 модели регрессии для обоих полов, в зависимости от города (Монреаль или Торонто) и места рождения (рожденный в Канаде или иммигрант). 

В конце концов, авторы пришли к следующим выводам:
1. Знание французского языка оказывает значимое влияние на занятость мигрантов Монреале. Этот фактор в значительной степени объясняет разницу в заработной плате мигрантов в Торонто и Монреале.
2. Страна происхождения оказывает влияние на то, насколько хорошо мигранты интегрируются в новую для них среду. Однако это не может объяснить различие в занятости мигрантов в Торонто и Монреале. 
3. Возможно, что иммигранты в Монреале подвергаются большей дискриминации, чем в Торонто. Знание французского или английского приносит меньший выигрыш мигрантам, чем местным, в Монреале и Торонто соответственно. 
4. Политика по отношению к мигрантам (больший акцент на знание французского) не является причиной для большего разрыва в занятости. Напротив, Монреаль привлекал бы таких же мигрантов, как и Торонто, и разрыв был бы еще больше.

### ["Economic Returns to Speaking the Right Language(s)? Evidence from Kazakhstan’s Shift in State Language and Language of Instruction", Alisher Aldashev, Alexander M. Danzer, (2014)](https://www.ssrn.com/abstract=2536267) 
Авторы проанализировали два больших опроса, один из которых был проведен до того, как казахский стал официальным государственным языком (Living Standard Measurement Survey 1996), а второй -- уже после (Research Institute CIOM multistage random sample, 2010). В обоих случаях знающие казахский получали меньший доход. Они также использовали инструменты -- количество лет, проведенных в городских районах, где живут в основном этнически казахи. Поскольку 2МНК-оценки оказались близки к оценкам обычного МНК, авторы пришли к выводу, что обратная причинно-следственная связь не является проблемой. Затем они обратились к данным PISA и пришли к выводу, что качество обучения на казахском хуже, чем на русском, для носителей обоих языков. Авторы связали это с нехваткой учебных материалов на казахском, низкими квалификацией и мотивацией учителей.

### Наиболее близкое к моему проекту исследование: [Ginsburgh, Victor A. and Prieto-Rodriguez, Juan (2011) "Returns to Foreign Languages of Native Workers in the EU"](http://ecares.ulb.ac.be/ecare/personal/ginsburgh/papers/170%20returns.pdf)

* анализировалось Mincer-type equation (также часть используется в исследованиях по оценке отачи от образования) с помощью OLS и IV quantile regression (Chernozhukov, Hansen). Оценки для OLS оказались существенно ниже. IV quantile regression, по мнению авторов, помогает снизить ненаблюдаемую гетерогенность, так как работники из одного квантиля наверняка обладают схожими ненаблюдаемыми характеристиками.

Знание языков повышают заработную плату индивида при прочих равных на 5-15%, в зависимости от страны. В меньшей степени в Дании и Нидерландах, где английский язык знают от 70 до 75 процентов населения.

* Результаты снижаются, если ввести контроль за профессией. 
* Спецификации приводят к похожим результатам.

# Заключение
В данном проекте были проанализированы данные RLMS для работающих индивидов 20 и 24 волны. 

Для индивидов, участвовавших в 24 волне, были получены значимые положительные МНК-оценки для английского, немецкого и "других" (испанский, португальский, латынь и японкий) языков. Однако с большой долей уверенности можно утверждать, что оценки в данных моделях страдают от смещения из-за ненаблюдаемых характеристик, а также ошибок измерения.   

Для данных 20 волны удалось подобрать набор релевантных и экзогенных инструментов для оценки влияния наличия высшего образования и знания иностранных языков. Оказалось, что эндогенность действительно присуствует: оценка для высшего образования оказалась завышенной, оценка для знания языков -- заниженной на части выборки, имеющей высшее образование. Если первый вывод кажется логичным, то насчет второго возникают вопросы. Такой результат (оценки с помощью инструментов значительно превышают оценки с помощью обычного МНК) был получены и в изученной мной литературе; ответ на данный вопрос дан, например, в [этой статье (стр. 601, Econometric Issues)](http://ecares.ulb.ac.be/ecare/personal/ginsburgh/papers/170%20returns.pdf). С одной стороны, при наличии ошибок измерения (сведения о владении языками дают сами индивиды без какого либо тестирования или проверки) стоит ожидать занижения оценок при МНК-оценке. С другой стороны, при оценке возникает проблема эндогенности из-за ability bias, то стоит ожидать завышения оценок обычным МНК. 

МНК-оценки для владения языками оказались выше для 24 волны, чем для 20. Однако из этого нельзя сделать однозначный вывод о том, что лингвистические способности стали цениться выше, так как 24 волна не была оценена с помощью инструментов.
